# OpenStreetMap für Webkarten

```python
import folium
from folium.plugins import MarkerCluster
```

Creating and programming of web maps and web apps with OpenStreetMap data and services.

To save a map as a .html save the map in a variable and invoke the method save("NameOfTheHTML.html") on it.

## Aufgabe 1 Eigene interaktive Webkarte mit Folium

```python
folium.Map(location=[47.3686, 8.5392], tiles='Stamen Toner', zoom_start=13)
```

## Aufgabe 2 Eigene interaktive Webkarte mit eigenen Geoobjekten

```python
aufgabe_2 = folium.Map(location=[47.22645, 8.81934],
               tiles='Stamen Toner',
               zoom_start=15)

folium.Marker(location=[47.22927, 8.82082],
              popup='Sonnenhof',
              icon=folium.Icon(color='green')
              ).add_to(aufgabe_2)

folium.Circle(radius=100,
              location=[47.22329, 8.81734],
              popup='Hochschule für Technik Rapperswil',
              color='blue',
              fill=True).add_to(aufgabe_2)
aufgabe_2
```

## Aufgabe 3 Eigene interaktive Webkarte mit statischem File (GeoJSON) einbinden

```python
aufgabe_3 = folium.Map(location=[47.3686, 8.5392],
               tiles='Stamen Toner',
               zoom_start=10)

folium.GeoJson(
     'castles.geojson',
     name='geojson'
 ).add_to(aufgabe_3)

folium.LayerControl().add_to(aufgabe_3)

aufgabe_3
```

## Aufgabe 4 Eigene interaktive Webkarte mit statischem File (GeoJSON) einbinden mit Point Clustering

```python
aufgabe_4 = folium.Map(location=[47.3686, 8.5392],
               tiles='Stamen Toner',
               zoom_start=8)

marker_cluster = MarkerCluster(
    overlay=True,
    control=True
)

marker_cluster.add_to(aufgabe_4)

folium.GeoJson(
     'castles.geojson',
     name='geojson'
 ).add_to(marker_cluster)

aufgabe_4
```
